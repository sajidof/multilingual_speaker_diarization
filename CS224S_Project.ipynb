# Data Setup

```python
import os
import glob

Countries = ["Australia", "Czech", "Hong_Kong", "Japan","Netherlands", "USA"]

for country in Countries:

  input_folder = '/content/drive/MyDrive/CS224S_Final_Project/data/TIMSS-Science/Videos/' + country
  output_folder = '/content/drive/MyDrive/CS224S_Final_Project/data/TIMSS-Science/Audio/' + country

  # Find all MP4 files in the input folder
  files = glob.glob(os.path.join(input_folder, '*.mp4'))

  for file in files:
      # Construct output file path with the same name but with WAV extension
      out_file = os.path.join(output_folder, os.path.basename(file)[:-4] + '.wav')

      # Convert MP4 to WAV using FFmpeg
      os.system(f'ffmpeg -i "{file}" -vn -ab 128k -ar 44100 -y "{out_file}"')
```
