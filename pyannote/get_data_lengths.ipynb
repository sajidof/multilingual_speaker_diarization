```python
import wave
import contextlib

def get_total_length(file_list_path):
    total_length_seconds = 0

    # Read the file names from the text file
    with open(file_list_path, 'r') as file:
        for line in file:
            filename = line.strip()
            print("Reading the file", filename)
            filename = filename + ".wav"

            try:
                with contextlib.closing(wave.open(filename, 'r')) as audio:
                    frames = audio.getnframes()
                    rate = audio.getframerate()
                    duration = frames / float(rate)
                    print(f"{filename} has length {duration:.2f} seconds")
                    total_length_seconds += duration
            except wave.Error as e:
                print(f"Error reading {filename}: {e}")

    total_minutes = int(total_length_seconds // 60)
    total_seconds = int(total_length_seconds % 60)
    return total_minutes, total_seconds
```
